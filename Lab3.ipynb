Task 1

```python
arr = [1, 5, 10, 50, 100, 500, 1000, 10000]
S = int(input())
cnt = 0
for i in reversed(arr):
  while S >= i:
    S -= i
    cnt += 1
print(cnt)
```

Task 2

```python
n = int(input())
sm = 0
for i in range(1, n + 1):
  mp = 1
  for j in range(i, 2 * i + 1):
    mp *= j
  sm += mp
print(sm)
```

Task 3

```python
import math

eps = 10 ** -3
a1 = 0
arr = [a1]
i = 1
while True:
  a2 = math.atan(a1) + 1
  arr.append(a2)
  i += 1
  if abs(a2 - a1) < eps:
    print(i)
    break
  a1 = a2
for a in arr:
  print(a, end=' ')
```

```python
k = int(input())
ai = int(input())
cnt = 0
while ai != 0:
  cnt += int(ai < k)
  ai = int(input())
print(cnt)
```
