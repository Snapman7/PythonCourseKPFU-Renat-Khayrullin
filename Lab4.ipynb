Task 2

```python
def lcm(a, b):
  if b == 0:
    return a
  return lcm(b, a % b)

def mpFraction(A, B, C, D):
  leastCommonMp = lcm(A * B, C * D)
  return (A * B // leastCommonMp, C * D // leastCommonMp)

A, B = list(map(int, input().split('/')))
C, D = list(map(int, input().split('/')))
E, F = mpFraction(A, B, C, D)
print(E, F, sep='/')
```

```python
def next_permutation(arr):
  sz = len(arr)
  i_first = -1
  for i in range(sz - 1, 0, -1):
    if arr[i - 1] < arr[i]:
      i_first = i - 1
      break
  if (i_first == -1):
    return (arr, False)
  for j in range(sz - 1, i_first, -1):
    if arr[j] > arr[i_first]:
      tmp = arr[j]
      arr[j] = arr[i_first]
      arr[i_first] = tmp
      break
  return (arr[0:(i_first + 1)] + list(reversed(arr[(i_first + 1):sz])), True)

n = int(input())
arr = list(map(int, input().split()))
newArr, hasNext = (sorted(arr), True)
while hasNext:
  print(newArr)
  newArr, hasNext = next_permutation(newArr)
```
