Task 1

```python
import matplotlib.pyplot as plt
import math
import numpy as np

def y(x):
  return (math.sin(float(x)) ** 2) - math.cos(float(x)) * 3 * float(math.e ** (-2 * float(x)))

ys = np.vectorize(y)
xs = np.linspace(-1, 1, 100)
plt.plot(xs, ys(xs))
plt.xlabel('x')
plt.ylabel('y')
plt.title('График функции')
plt.grid()

plt.show()
```

```python
x = float(input())
y = math.sin(x) * math.sin(x) - math.cos(x) * 3 * (math.e ** (-2 * x))
print(y)
```

```python
xs = np.linspace(-0.5, 0.5, 100)
zs = (math.e ** (9 * xs)) + 3 * (math.e ** (-2 * xs))
plt.plot(xs, zs)
plt.xlabel('x')
plt.ylabel('z')
plt.title('График функции')
plt.grid()

plt.show()
```
